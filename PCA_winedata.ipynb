```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

### Import data set

```python
wine=pd.read_csv('Wine.csv')
wine
```

```python
wine.describe()
```

```python
wine.info()
```

### considering numerical columns

```python
wine1=wine.iloc[:,1:]
wine1
```

```python
# converting into numpy array
wine1_array=wine1.values
wine1_array
```

### Normalization

```python
wine_normal=scale(wine1_array)
```

```python
wine_normal
```

## PCA 

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
#loadings or weights
pca.components_
```

```python
#amount of variance for each pca
var = pca.explained_variance_ratio_
var
```

```python
#cummulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
#Variance plot for PCA components obtained
plt.plot(var1,color="green")
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']), wine[['Type']]], axis = 1)
finalDf
```

```python
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=finalDf)
```

```python
pca_values=pca_values[:,0:3]
pca_values
```

## Heirarchical clustering 

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch  
z = linkage(pca_values, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
z,
)
plt.show()     
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=3,affinity = "euclidean").fit(wine_normal) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
wine['hclust']=cluster_labels 
wine
```

## K-Means clustering

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(wine_normal)
    WCSS.append(clf.inertia_) 
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
clf = KMeans(n_clusters=3)    
y_kmeans = clf.fit_predict(wine_normal)  
```

```python
y_kmeans
```

```python
md=pd.Series(y_kmeans)
wine['kclust']=md 
wine
```
